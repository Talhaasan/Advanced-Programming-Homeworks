```python
import pandas as pd
import numpy as np

print('a.')
automobiles = pd.read_excel('C:\\Users\\pc\\.spyder-py3\\Automobiles.xlsx',
                     usecols=['company','body-style','wheel-base','length','engine-type','num-of-cylinders','horsepower','average-mileage','price'],index_col=0)
print("first 10 rows:")
automobiles.head(10)
```

```python
print('b.')
dfb = automobiles
print(dfb)
automobilesDataFrame = input("Enter selected series :")
print("Content of the series:")
automobilesDataFrame = dfb.loc[automobilesDataFrame]
automobilesDataFrame
```

```python
print('c.')
dfc = automobiles
print("Create a new Company-Price")
company = input("Enter company :")
bodyStyle = input("Enter body-style :")
wheelBase = float(input("Enter wheel-base :"))
length = float(input("Enter length :"))
engineType = input("Enter engine-type:")
numOfCylinders = int(input("Enter num-of-cylinders :"))
horsepower = int(input("Enter horsepower :"))
avarageMileage = int(input("Enter average-mileage :"))
price = int(input("Enter price:"))
newCarCompany = [bodyStyle,wheelBase,length,engineType,numOfCylinders,horsepower,avarageMileage,price]
dfc.loc[company] = newCarCompany
print("Updated dataframe(adding new Car Company) :")
dfc
```

```python
print('d.')
dfd = automobiles
dfd = dfd.drop(['body-style','engine-type','length'],axis=1)
print()
print("Updated dataframe(deleting columns) :")
dfd
```

```python
print('e.')
dfe = automobiles
print("Ascending order :")
ascendingResult = dfe.price.sort_values(ascending=True)
print(ascendingResult)
print()
print("Descending order :")
descendingResult = dfe.price.sort_values(ascending=False)
descendingResult
```

```python
print('f.')

dff = automobiles
carBrand = dff.groupby('company',as_index=False)

IsuzuDf = carBrand.get_group('isuzu')
PorscheDf = carBrand.get_group('porsche')
print(IsuzuDf)
print(PorscheDf)

avarageIsuzuPrice = 6785
avaragePorschePrice =((34028+37028)/2)

IsuzuDf = IsuzuDf.replace(np.nan, avarageIsuzuPrice)
PorscheDf = PorscheDf.replace(np.nan, avaragePorschePrice)
IsuzuDf
PorscheDf


print()
dff.drop(dff.index['company' == 'isuzu'], inplace = True)
dff.drop(dff.index['company' == 'porsche'], inplace = True)

dataFrame = [IsuzuDf, PorscheDf,dff]
resultDF = pd.concat(dataFrame)

avaragePrice = resultDF["price"].mean()
print ('Average price :',avaragePrice)
```
