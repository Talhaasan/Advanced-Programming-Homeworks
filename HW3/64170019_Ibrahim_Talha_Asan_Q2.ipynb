```python
import pandas as pd
import matplotlib.pyplot as plt

print('a.')
dataframe = pd.read_excel('C:\\Users\\pc\\.spyder-py3\\COE-101-GRADEs.xlsx')
print("dataframe:")
dataframe
```

```python
#b.
dataframeWithAvarageCalculation = pd.DataFrame(0.25*dataframe['MIDI']+0.25*dataframe['MIDII']+0.5*dataframe['FIN'],columns=['AVERAGE'])
dataframe = dataframe.join(dataframeWithAvarageCalculation)
```

```python
#c.
dataframe["MIDI"].fillna(0)
dataframe["MIDII"].fillna(0)
dataframe["FIN"].fillna(0)
dataframe["AVERAGE"].fillna(0)
```

```python
print('d.')
GRADE=[]
for NumberArg in dataframe['AVERAGE']:
    if NumberArg>=94:
        GRADE.append('A+')
    elif NumberArg>=87:
        GRADE.append('A')
    elif NumberArg>=79:
        GRADE.append('B+')
    elif NumberArg>=70:
        GRADE.append('B')
    elif NumberArg>=60:
        GRADE.append('C+')
    elif NumberArg>=50:
        GRADE.append('C')
    elif NumberArg>=45:
        GRADE.append('D+')
    elif NumberArg>=40:
        GRADE.append('D')
    else:
        GRADE.append('F')

dataframeWithGRADE = pd.DataFrame(GRADE,columns=['GRADE'])
dataframe = dataframe.join(dataframeWithGRADE)
print("dataframe with GRADE:")
dataframe
```

```python
print('e. and f.')
failedCourse=0
passedCourse=0
for GRADE in dataframe['GRADE']:
    if GRADE=='D+' or GRADE=='D' or GRADE =='F':
        failedCourse+=1
    else:
        passedCourse+=1
print("Number of failedCourse in this course :",failedCourse)
print("Number of passedCourse in this course :",passedCourse)
```

```python
print('g.')
pivotTable = dataframe.pivot_table(index =['NAME'],values =['GRADE'],aggfunc ='sum')
pivotTable
```

```python
#f.
plt.hist(dataframe['GRADE'])
plt.title('Count of Students That Took Grade (Histogram Graph)')
```

```python
#g.
fig = plt.figure()
ax = fig.add_subplot()
dataframe.GRADE.value_counts().plot(kind='pie')
plt.axis('equal')
plt.title('Count of Students That Took Grade (Pie Graph)')
```

```python
print('h.')
MIDI = dataframe['MIDI']
MIDII = dataframe['MIDII']
FIN = dataframe['FIN']
correlation1 = MIDI.corr(MIDII)
correlation2 = MIDII.corr(FIN)
print('Correlation value between MIDI and MIDII :',correlation1)
print('Correlation value between MIDI and FIN :',correlation2)
print('Result Interpretation:')
print('Since the correlation value between MIDII and FIN is higher than the correlation value between MIDI and MIDII.')
print('In a conclusion MIDII and FIN  have a higher relationship.')
```

```python
print('i.')
MIDI_mean = dataframe["MIDI"].mean()
print ('Average of MIDI grades :',MIDI_mean)
MIDII_mean = dataframe["MIDII"].mean()
print ('Average of MIDII grades :',MIDII_mean)
print()
increased_FINALtoMIDI = dataframe[dataframe['FIN'] > ((MIDI_mean*110)/100)]
print('Students whose final grades increased more %10 than the average of MIDI')
print(increased_FINALtoMIDI)
print()
print('Students whose final grades increased more %10 than the average of MIDII')
increased_FINALtoMIDII = dataframe[dataframe['FIN'] > ((MIDII_mean*110)/100)]
print(increased_FINALtoMIDII)
```
