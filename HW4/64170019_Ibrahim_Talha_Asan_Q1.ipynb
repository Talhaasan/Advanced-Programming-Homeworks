```python
from sklearn import datasets
from sklearn import model_selection
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

boston = datasets.load_boston()
x = boston.data;
y = boston.target;
x_test,x_train,y_test,y_train= model_selection.train_test_split(x,y,random_state = 0);
```

```python
ridge_regression = Ridge(alpha=0.0001)
ridge_regression.fit(x_train, y_train)
y_pred_default = ridge_regression .predict(x_test)
print("Ridge Regression Test Score :",ridge_regression .score(x_test,y_test))
plt.scatter(y_test,y_pred_default)
plt.title('Scatter Plot')
plt.xlabel('true value')
plt.ylabel('predicted value')
plt.show()
```

```python
ridge_regression_default_biggerAlpha = Ridge(alpha=10000)
ridge_regression_default_biggerAlpha.fit(x_train, y_train)
y_prediction_default = ridge_regression_default_biggerAlpha.predict(x_test)
print("Ridge Regression Test Score :",ridge_regression_default_biggerAlpha.score(x_test,y_test))
plt.scatter(y_test,y_prediction_default)
plt.title('Scatter Plot')
plt.xlabel('true value')
plt.ylabel('predicted value')
plt.show()
```

```python
polynomial_features = PolynomialFeatures(2)
polynomial_features.fit(x_train)
x_train = polynomial_features.transform(x_train)
x_test = polynomial_features.transform(x_test)
x_test.shape,x_train.shape
```

```python
ridge_regression_Polynomial = Ridge(alpha=10000)
ridge_regression_Polynomial.fit(x_train, y_train)
y_pred_poly = ridge_regression_Polynomial.predict(x_test)
print("Ridge Regression(bigger alpha value) Test Score :",ridge_regression_Polynomial.score(x_test,y_test))
plt.title('Scatter Plot')
plt.xlabel('true value')
plt.ylabel('predicted value')
plt.scatter(y_test,y_pred_poly)
plt.show()
```

```python
ridge_regression_Polynomial_biggerAlpha = Ridge(alpha=10000000)
ridge_regression_Polynomial_biggerAlpha.fit(x_train, y_train)
y_prediction_poly = ridge_regression_Polynomial_biggerAlpha.predict(x_test)
print("Ridge Regression(bigger alpha value) Test Score :",ridge_regression_Polynomial_biggerAlpha.score(x_test,y_test))
plt.title('Scatter Plot')
plt.xlabel('true value')
plt.ylabel('predicted value')
plt.scatter(y_test,y_prediction_poly)
plt.show()
```

# Results

## In different alpha and feature trials, the best test score was obtained when the original attributes and alpha value were equal to 0.0001.
